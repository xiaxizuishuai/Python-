<font size="5">查看数据的编码</font>
>注意路径，要用‘/’，而不能用‘\’
>>使用相对路径，一定要和代码在同一个文件夹下面;相对路径还利于小组成员查看，合作

```python
import chardet

def check_encoding(filename):
    rawdata = open(filename, 'rb').read()
    result = chardet.detect(rawdata)
    encoding = result['encoding']
    confidence = result['confidence']
    return encoding, confidence

file_path = './000001.csv'# 我用相对路径./表示当前目录
encoding, confidence = check_encoding(file_path)
print(f"Encoding: {encoding}, Confidence: {confidence}")
```

<font size="5" color=pink>将'Day'列从字符串格式转换为日期时间格式</font>

```python
data['Day'] = pd.to_datetime(data['Day'],format = '%Y/%m/%d') 
data
# pd.to_datetime() 将字符串转换为datetime对象
# format='%Y/%m/%d' 指定了输入日期的格式：年/月/日（如2025/9/10）
```

```python
data = data.sort_values(by=['Day'], axis=0, ascending=True) #对'Day'列降序排序，axis=0表示列，1为行，True为升序，False为降序
data
```

<font size="6"> 更改行索引 </font>   
<font size="5" color=pink>注意</font>   
inplace=False/True   
False原数据框上**不会**进行修改  
True直接在原数据框上进行修改，不创建新的数据框  

```python
df2=data.set_index('Day', inplace =False)
df2
data
```

```python
#改'Day'为行索引，这样更改后可直接通过日期访问行数据
data.set_index('Day', inplace = True)
data #inplace=True 表示直接在原数据框上进行修改，不创建新的数据框
```
